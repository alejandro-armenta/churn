```python
import pandas as pd
import numpy as np

churn_data = pd.read_csv('socialnet7_dataset2.csv', index_col=[0,1])
churn_data
```

```python
data_scores = churn_data.copy()
data_scores = data_scores.drop('is_churn', axis=1)
data_scores
```

```python
stats = pd.read_csv('stats.csv', index_col=0)
stats
```

```python
stats = stats.drop('is_churn', axis=0)
```

```python
stats
```

```python
#estas son las normales
skew_threshold = 4.0
skewed_columns = (stats['skew'] > skew_threshold) & (stats['min'] >= 0.0)
skewed_columns = skewed_columns[skewed_columns]
skewed_columns
```

```python
for col in skewed_columns.keys():
    data_scores[col] = np.log(1.0 + data_scores[col])
```

```python
data_scores
```

```python
fattail_columns = (stats['skew'] > skew_threshold) & (stats['min'] < 0.0)

for col in fattail_columns[fattail_columns].keys():
    
    pass
```

HASTA AQUI

```python
for col in skewed_columns.keys():
    data_scores[col] = np.log(data_scores[col] + 1.0)
    stats.at[col, 'mean'] = data_scores[col].mean()
    stats.at[col, 'std'] = data_scores[col].std()
```

```python
data_scores = (data_scores-stats['mean'])/stats['std']
```

```python
data_scores['is_churn'] = churn_data['is_churn']
```

```python
data_scores
```

```python
data_scores.to_csv('scores.csv',header=True)
```
