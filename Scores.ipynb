```python
import pandas as pd
import numpy as np

churn_data = pd.read_csv('socialnet7_dataset.csv', index_col=[0,1])
churn_data
```

```python
data_scores = churn_data.copy()
data_scores.drop('is_churn', axis=1)
```

```python
stats = pd.read_csv('stats.csv', index_col=0)
stats = stats.drop('is_churn', axis=0)

skew_threshold = 4.0
skewed_columns = (stats['skew'] > skew_threshold) & (stats['min'] >= 0.0)
skewed_columns = skewed_columns[skewed_columns]
skewed_columns
```

```python
for col in skewed_columns.keys():
    data_scores[col] = np.log(data_scores[col] + 1.0)
    #estos son mu
    stats.at[col, 'mean'] = data_scores[col].mean()
    stats.at[col, 'std'] = data_scores[col].std()
    
    #print(col, data_scores[col].mean())
    #print(col, data_scores[col].std())
```

```python
data_scores = (data_scores-stats['mean'])/stats['std']
```

```python
data_scores['is_churn'] = churn_data['is_churn']
```

```python
data_scores.to_csv('scores.csv',header=True)
```
